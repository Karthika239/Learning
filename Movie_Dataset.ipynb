```python
# Supress Warnings

import warnings
warnings.filterwarnings('ignore')

# Import the numpy and pandas packages

import numpy as np
import pandas as pd
```

```python
#Read Dataset
movies = pd.read_csv('/content/Movie_Data.csv')
display(movies.head())
```

```python
#Column-wise null count
movies.isnull().sum()
```

```python
#Row-wise null count :
movies.isnull().sum(axis=1)
```

```python
#Column-wise null percentages:

round(movies.isnull().sum() / len(movies) * 100, 2)
```

How it works:

isnull() — returns a DataFrame of True/False for each value

.sum() — counts True values (column-wise by default)

.sum(axis=1) — counts across columns (row-wise)

/ len(movies) * 100 — converts count to percentage

round(..., 2) — rounds to 2 decimal places

```python
#Drop unnecessary columns
columns_to_drop = [
    'color', 'director_facebook_likes', 'actor_1_facebook_likes',
    'actor_2_facebook_likes', 'actor_3_facebook_likes', 'actor_2_name',
    'cast_total_facebook_likes', 'actor_3_name', 'duration',
    'facenumber_in_poster', 'content_rating', 'country',
    'movie_imdb_link', 'aspect_ratio', 'plot_keywords'
]

movies = movies.drop(columns=columns_to_drop)
```

```python
movies.head()
```

How it works:

drop(columns=...) — removes the listed columns

movies = movies.drop(...) — reassigns so the change is saved back to movies

movies.head() — lets you inspect the result to confirm columns are gone

```python
#Drop rows where columns have >5% null values:
null_percent = round(movies.isnull().sum() / len(movies) * 100, 2)
high_null_cols = null_percent[null_percent > 5].index
```

```python
movies = movies.dropna(subset=high_null_cols)
```

```python
movies.shape
```

How it works:

null_percent — reuses the percentage calculation from subtask 2.1

null_percent > 5 — filters to only columns where nulls exceed 5%

.index — extracts just the column names

dropna(subset=high_null_cols) — drops any row that has a NaN in any of those high-null columns

movies.shape — lets you verify how many rows remain after dropping

```python
#drop unnecessary rows
movies = movies[movies.isnull().sum(axis=1) <= 5]
movies.shape
```

```python
#Fill NaN values in the language column with 'English'
movies['language'].fillna('English', inplace=True)
movies['language'].isnull().sum()
```

How it works

fillna('English') — replaces all NaN values in the language column with 'English'

inplace=True — modifies the dataframe directly without needing to reassign

.isnull().sum() — confirms the result should be 0 (no nulls left)

Perform Data Analysis on this Clean data:


```python
#Change unit of columns
#Convert budget and gross from $ to million $:

movies['budget'] = movies['budget'] / 1000000
movies['gross'] = movies['gross'] / 1000000

movies[['budget', 'gross']].head()
```

How it works:

Dividing by 1000000 converts the values from dollars to million dollars

movies[['budget', 'gross']].head() — inspect both columns to confirm the conversion worked

```python
#Movies with highest Profit:
#Create the profit column:
movies['profit'] = movies['gross'] - movies['budget']
movies.head()
```

```python
#Sort the dataframe by profit:
movies = movies.sort_values('profit', ascending=False)
movies.head()
```

```python
#Extract top 10 profiting movies:
top10 = movies.head(10)
top10[['movie_title', 'director_name', 'budget', 'gross', 'profit']]
```

How it works:

gross - budget — profit is simply revenue minus cost

sort_values('profit', ascending=False) — sorts in descending order (highest profit first)

movies.head(10) — extracts the top 10 rows after sorting

```python
#drop duplicates
movies = movies.drop_duplicates()
movies.shape
```

```python
#recalculate:
movies['profit'] = movies['gross'] - movies['budget']
movies = movies.sort_values('profit', ascending=False)

top10 = movies.head(10)
top10[['movie_title', 'director_name', 'budget', 'gross', 'profit']]
```

How it works:

drop_duplicates() — removes all duplicate rows from the dataframe

After dropping duplicates, we redo the profit calculation and sorting to get a clean top 10

```python
# TOP 250 movies
IMDb_Top_250 = movies[movies['num_voted_users'] > 25000].sort_values('imdb_score', ascending=False).head(250)
```

```python
IMDb_Top_250['Rank'] = range(1, 251)

IMDb_Top_250[['Rank', 'movie_title', 'director_name', 'imdb_score', 'num_voted_users']]
```

```python
# Extract top foreign language films:
```

```python
Top_Foreign_Lang_Film = IMDb_Top_250[IMDb_Top_250['language'] != 'English']
Top_Foreign_Lang_Film[['Rank', 'movie_title', 'director_name', 'language', 'imdb_score']]
```

How it works:

num_voted_users > 25000 — filters out movies with too few votes (ensures credibility)

sort_values('imdb_score', ascending=False) — highest rated first

.head(250) — keeps only top 250
range(1, 251) — assigns ranks 1 to 250

language != 'English' — filters to only non-English movies

```python
#Extract top 10 directors by mean IMDb score:
top10director = movies.groupby('director_name')['imdb_score'].mean().sort_values(ascending=False).head(10).reset_index()

top10director.columns = ['director_name', 'mean_imdb_score']
top10director
```

How it works:

groupby('director_name') — groups all movies by director

['imdb_score'].mean() — calculates the mean IMDb score for each director

sort_values(ascending=False) — sorts highest mean score first

.head(10) — keeps only top 10

reset_index() — converts the result back to a clean dataframe
.columns — renames columns for clarity

```python
# find popular genre
movies['genre_1'] = movies['genres'].apply(lambda x: x.split('|')[0])
movies['genre_2'] = movies['genres'].apply(lambda x: x.split('|')[1] if len(x.split('|')) > 1 else x.split('|')[0])

movies[['movie_title', 'genres', 'genre_1', 'genre_2']].head()
```

```python
#Group the dataframe:
movies_by_segment = movies.groupby(['genre_1', 'genre_2'])
```

```python
#5 most popular genre
PopGenre = movies_by_segment['gross'].mean().sort_values(ascending=False).head(5).reset_index()

PopGenre.columns = ['genre_1', 'genre_2', 'mean_gross']
PopGenre
```

```python
#find critic favourite and audience favourite actors
Meryl_Streep = movies[movies['actor_1_name'] == 'Meryl Streep']
Meryl_Streep
```

```python
Leo_Caprio = movies[movies['actor_1_name'] == 'Leonardo DiCaprio']
Leo_Caprio
```

```python
Brad_Pitt = movies[movies['actor_1_name'] == 'Brad Pitt']
Brad_Pitt
```

```python
Combined = pd.concat([Meryl_Streep, Leo_Caprio, Brad_Pitt])
Combined
```

```python
Combined_by_actor = Combined.groupby('actor_1_name')
```

```python
Combined_by_actor[['num_critic_for_reviews', 'num_user_for_reviews']].mean()
```

How it works:

actor_1_name == '...' — filters movies where the actor is the lead

pd.concat([...]) — appends all three dataframes row-wise into one

groupby('actor_1_name') — groups by actor

.mean() — calculates mean for both review columns
